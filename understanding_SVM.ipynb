```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Create toy dataset
X, y = datasets.make_moons(n_samples=100, noise=0.1, random_state=42)

# Try different kernels
kernels = ['linear', 'poly', 'rbf']
models = [SVC(kernel=k).fit(X, y) for k in kernels]

# Plot decision boundaries
for i, (model, k) in enumerate(zip(models, kernels), 1):
    plt.subplot(1, 3, i)
    plt.title(k)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
    
    # decision boundary
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100),
                         np.linspace(ylim[0], ylim[1], 100))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate circular data (inside vs outside circle)
np.random.seed(0)
n = 100
X = np.random.uniform(-1, 1, (n, 2))
y = (X[:,0]**2 + X[:,1]**2 > 0.5).astype(int)  # inside circle=0, outside=1

# Original 2D plot
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("2D Space (Not Separable)")
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.xlabel("x1")
plt.ylabel("x2")
plt.gca().set_aspect("equal", adjustable="box")

# Transform to 3D: add z = x^2 + y^2
from mpl_toolkits.mplot3d import Axes3D
fig = plt.subplot(1,2,2, projection='3d')
z = X[:,0]**2 + X[:,1]**2
fig.scatter(X[:,0], X[:,1], z, c=y, cmap="coolwarm")
fig.set_title("3D Space with z = x1^2 + x2^2")
fig.set_xlabel("x1")
fig.set_ylabel("x2")
fig.set_zlabel("z")

plt.tight_layout()
plt.show()
```
