```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

y_resampled.value_counts()
```

```python
train_X, test_X, train_y, test_y = train_test_split(X_resampled, y_resampled, random_state=42)
```

```python
GradBoostclf2 = GradientBoostingClassifier()

GradBoostclf2.fit(train_X, train_y)
```

```python
GradBoostPred2 = GradBoostclf2.predict(test_X)

print("accuracy score:",accuracy_score(test_y, GradBoostPred2))
print("RMSE:", root_mean_squared_error(test_y, GradBoostPred2))
print("Mean Absolute Error:", mean_absolute_error(test_y, GradBoostPred2))
```

```python
RandomFor2 = RandomForestClassifier()

RandomFor2.fit(train_X, train_y)
```

```python
RandForPred2 = RandomFor2.predict(test_X)

print("accuracy score:",accuracy_score(test_y, RandForPred2))
print("RMSE:", root_mean_squared_error(test_y, RandForPred2))
print("Mean Absolute Error:", mean_absolute_error(test_y, RandForPred2))
```

```python
RandomFor3 = RandomForestClassifier(n_estimators=75, max_depth=75)

RandomFor3.fit(train_X, train_y)
```

```python
RandForPred3 = RandomFor3.predict(test_X)

print("accuracy score:",accuracy_score(test_y, RandForPred3))
print("RMSE:", root_mean_squared_error(test_y, RandForPred3))
print("Mean Absolute Error:", mean_absolute_error(test_y, RandForPred3))
```
